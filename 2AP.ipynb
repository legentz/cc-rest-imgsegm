```python
#disegna la curva rappresentata da y = 3x^2 + 2x + 10
import matplotlib.pyplot as plt

x = []
y = []

x_asc = [-100, 100]
y_asc = [0, 0]

x_ord = [0,0]
y_ord = [0, 30000]


for i in range(-100,101):
    x.append(i)
    y.append(3*(i**2) + 2 * i + 10)

plt.plot(x,y, color='red', linestyle='-', linewidth=2)
plt.plot(x_asc, y_asc, color='green', linestyle='-.', linewidth=1)
plt.plot(x_ord, y_ord, color='black', linestyle='-.', linewidth=1)
plt.title("grafico della funzione y = 3x^2 + 2x + 10")
plt.xlabel("Asse X")
plt.ylabel("Asse Y")
plt.show()
```

```python
#Simula il lancio di un dado 200 volte (numeri da 1 a 6) e visualizza in un istogramma quante volte è uscito ciascun numero.
import random

etichette = ["uno", "due", "tre", "quattro", "cinque", "sei"]
numeri = [0,0,0,0,0,0]

for _ in range(10000):
    generato = random.randint(1, 6)
    numeri[generato - 1] = numeri[generato - 1] + 1


plt.bar(etichette, numeri)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

capitale = 100
andamento = [capitale]

for _ in range(100):
    vittoria = random.randint(0,1)
    if vittoria == 0:
        capitale = capitale - capitale * 0.5
    else:
        capitale = capitale + capitale * 0.5
    andamento.append(capitale)
    if capitale < 10:
        break

plt.plot(andamento)
plt.title("100 giorni di scomesse")
plt.xlabel("giorni")
plt.ylabel("andamento soldi")
plt.show()
```
